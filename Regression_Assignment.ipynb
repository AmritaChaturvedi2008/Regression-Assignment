# Regression Assignment

1. What is Simple Linear Regression?

Answer:

Simple Linear Regression is a method to find the relationship between two variables: one independent variable (input) and one dependent variable (output). It tries to fit a straight line through the data points to predict the output based on the input.


---

2. What are the key assumptions of Simple Linear Regression?


Answer:

The relationship between variables is linear.

The residuals (errors) are normally distributed.

The residuals have constant variance (homoscedasticity).

There is little or no multicollinearity.

The residuals are independent (no autocorrelation).



---

3. What does the coefficient m represent in the equation Y = mX + c?


Answer:
The coefficient m represents the slope of the line. It shows how much Y changes for a one-unit change in X.


---

4. What does the intercept c represent in the equation Y = mX + c?

Answer:

The intercept c is the value of Y when X is 0. It shows where the line crosses the Y-axis.


---

5. How do we calculate the slope m in Simple Linear Regression?

Answer:

The slope m is calculated using the formula:

m = \frac{\sum (X - \bar{X})(Y - \bar{Y})}{\sum (X - \bar{X})^2}


---

6. What is the purpose of the least squares method in Simple Linear Regression?

Answer:

The least squares method tries to find the line that minimizes the sum of squared errors (the vertical distance between the points and the line).


---

7. How is the coefficient of determination (R²) interpreted in Simple Linear Regression?

Answer:

R² shows how well the independent variable explains the dependent variable.

If R² = 1, it means perfect prediction.

If R² = 0, it means the model explains nothing.



---

8. What is Multiple Linear Regression?

Answer:

Multiple Linear Regression is a method to find the relationship between two or more independent variables and one dependent variable.


---

9. What is the main difference between Simple and Multiple Linear Regression?

Answer:

Simple Linear Regression: Only one independent variable.

Multiple Linear Regression: Two or more independent variables.



---

10. What are the key assumptions of Multiple Linear Regression?

Answer:

Linear relationship between dependent and independent variables.

Multivariate normality (errors are normally distributed).

No multicollinearity (independent variables are not too highly correlated).

Homoscedasticity (constant variance of errors).

Independence of errors.



---

11. What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model?

Answer:

Heteroscedasticity means the errors have different variances at different levels of the independent variables.
It makes the model's predictions unreliable and affects the accuracy of statistical tests.


---

12. How can you improve a Multiple Linear Regression model with high multicollinearity?

Answer:

Remove highly correlated variables.

Use techniques like Principal Component Analysis (PCA).

Combine correlated variables.

Use regularization methods like Ridge or Lasso regression.



---

13. What are some common techniques for transforming categorical variables for use in regression models?

Answer:

One-Hot Encoding: Creating separate binary columns for each category.

Label Encoding: Giving each category a unique number.

Dummy Variables: Similar to one-hot encoding but one less column.



---

14. What is the role of interaction terms in Multiple Linear Regression?
Answer:

Interaction terms show how the effect of one independent variable on the dependent variable depends on another variable.


---

15. How can the interpretation of intercept differ between Simple and Multiple Linear Regression?

Answer:

In Simple Linear Regression, intercept is the value of Y when X = 0.

In Multiple Linear Regression, intercept is the value of Y when all independent variables are 0.



---

16. What is the significance of the slope in regression analysis, and how does it affect predictions?

Answer:

The slope shows the rate of change between the independent and dependent variables.
A higher slope means a bigger effect on the prediction.


---

17. How does the intercept in a regression model provide context for the relationship between variables?

Answer:

The intercept helps to understand the starting point of Y when all independent variables are 0, providing a base reference for predictions.


---

18. What are the limitations of using R² as a sole measure of model performance?

Answer:


R² can be high even for a bad model.

It doesn’t tell about overfitting.

It does not indicate if the model is biased.

It doesn’t work well with non-linear relationships.



---

19. How would you interpret a large standard error for a regression coefficient?

Answer:

A large standard error means the coefficient is not estimated accurately, and the true effect might be different from the estimated effect.


---

20. How can heteroscedasticity be identified in residual plots, and why is it important to address it?

Answer:

In residual plots, heteroscedasticity appears as a fanning out or funnel shape of errors.
It’s important to fix because it can make predictions and confidence intervals wrong.


---

21. What does it mean if a Multiple Linear Regression model has a high R² but low adjusted R²?

Answer:

It means the model might have too many useless variables that do not really help in prediction.


---

22. Why is it important to scale variables in Multiple Linear Regression?
Answer:

Scaling ensures that variables are on the same scale, which helps in faster learning and more accurate models, especially when using regularization.


---

23. What is Polynomial Regression?

Answer:

Polynomial Regression is a type of regression where the relationship between the independent variable and the dependent variable is curved instead of straight.


---

24. How does Polynomial Regression differ from Linear Regression?

Answer:


Linear Regression fits a straight line.

Polynomial Regression fits a curved line (like a parabola).



---

25. When is Polynomial Regression used?

Answer:

Polynomial Regression is used when the data shows a non-linear relationship but we still want to use a regression model.


---

26. What is the general equation for polynomial regression?

Answer:

The general equation for polynomial regression is:

Y = b_0 + b_1X + b_2X^2 + b_3X^3 + \ldots + b_nX^n

28. What are the limitations of polynomial regression?

Answer:

It can easily overfit the data if the degree is too high.

It becomes complex and hard to interpret.

Extrapolation (predicting outside the data range) can give wrong results.

Needs more computation for higher degrees.



---

29. What methods can be used to evaluate model fit when selecting the degree of a polynomial?

Answer:

Use Cross-Validation to check model performance.

Look at R² and Adjusted R² values.

Check Mean Squared Error (MSE) or Root Mean Squared Error (RMSE).

Use visual plots to see how well the curve fits the data.



---

30. Why is visualization important in polynomial regression?

Answer:

Visualization helps to:

See the shape of the curve fitting the data.

Detect overfitting or underfitting.

Understand how well the model captures the relationship between variables.
